```python
import pandas as pd
from IPython.display import display
pd.set_option('display.max_columns', None) #Para ver todas as colunas
url = r"C:\Users\Gioher\Downloads\archive (1)\V_OCORRENCIA_AMPLA.json"
df = pd.read_json(url, encoding='UTF-8-SIG')
df.head(3)
```

```python
#SCRIPT PARA LER A CRIPTOGRAFIA DO ARQUIVO JSON
#pip install chardet
#pip install codecs
import codecs
from chardet.universaldetector import UniversalDetector

caminho_do_arquivo = r"C:\Users\edmil\OneDrive\Documentos\Cursos\Python - SGBDS - Arquivos\Origem de dados\V_OCORRENCIA_AMPLA.json"
detector = UniversalDetector()

with codecs.open(caminho_do_arquivo, 'rb') as arquivo: #rb esta lendo como binario em forma de bits no lugar de texto ou outra coisa 
    for linha in arquivo:
        detector.feed(linha)
        if detector.done:
            break

detector.close()
encoding_detectado = detector.result['encoding']

print(encoding_detectado) #
```
